```python
from amazon_scraper import AmazonScraper
from lxml import html
```

```python
scraper = AmazonScraper()
```

```python
# page = scraper.get_html('https://www.amazon.com/dp/B07C3R49QR')
# page = scraper.get_html('https://www.amazon.com/dp/B011PHGURS')
page = scraper.get_html('https://www.amazon.com/dp/B01N4AODB3')
doc = html.fromstring(page.content)
```

```python
# XPATH_PROD_DESC = '//*[@id="aplus"]/descendant-or-self::*[img]'
XPATH_PROD_DESC_IMG = '//*[@id="aplus"]//img'
```

```python
doc.xpath(XPATH_PROD_DESC_IMG)
```

```python
for item in doc.xpath(XPATH_PROD_DESC_IMG):
    print(item.xpath('@src'))
```

```python
XPATH_PROD_DESC_TEXT = '//*[@id="aplus"]/descendant-or-self::*/text()'
```

```python
for item in doc.xpath(XPATH_PROD_DESC_TEXT):
    if 'div' not in str(item) and 'align' not in str(item) and 'execute' not in str(item):
        if str(item).strip() and 'Read more' not in str(item).strip():
            print(str(item).strip())
            print('-------------------------')
```
